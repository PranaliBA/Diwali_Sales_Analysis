# Exploratory Data Analysis

### Gender

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Standardize Gender values
df['Gender'] = df['Gender'].replace({'M': 'Male', 'F': 'Female'})

# Create a larger figure
fig, ax = plt.subplots(figsize=(10, 6))

# Define the correct palette based on Gender values
colors = {'Male': '#1f77b4', 'Female': '#ff7f0e'}

# Create the countplot with hue and disable the legend
sns.countplot(x='Gender', data=df, ax=ax, hue='Gender', palette=colors, width=0.6, legend=False)

# Add bar labels
for bars in ax.containers:
    ax.bar_label(bars, fmt='%d', fontsize=12, label_type='edge', color='black', padding=3)

# Customize the appearance
ax.set_title('Gender Distribution', fontsize=16)
ax.set_xlabel('Gender', fontsize=14)
ax.set_ylabel('Count', fontsize=14)
ax.tick_params(axis='both', labelsize=12)

# Show plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Ensure the Gender column contains 'M' and 'F' instead of 'Male' and 'Female'
df['Gender'] = df['Gender'].replace({'Male': 'M', 'Female': 'F'})

# Aggregate total sales amount by Gender
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

# Create a larger figure
fig, ax = plt.subplots(figsize=(10, 6))

# Define color palette corresponding to 'M' and 'F'
colors = {'M': '#1f77b4', 'F': '#ff7f0e'}

# Generate bar plot with assigned colors
ax = sns.barplot(x='Gender', y='Amount', data=sales_gen, hue='Gender', palette=colors, width=0.6, legend=False)

# Set plot title and labels for clarity
ax.set_title('Total Sales Amount by Gender', fontsize=16)
ax.set_xlabel('Gender', fontsize=14)
ax.set_ylabel('Total Amount', fontsize=14)
ax.tick_params(axis='both', labelsize=12)

# Display the plot
plt.show()
```

*From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men*

### Age

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Ensure Gender values remain as 'M' and 'F'
df['Gender'] = df['Gender'].replace({'Male': 'M', 'Female': 'F'})

# Create a larger figure
fig, ax = plt.subplots(figsize=(10, 6))

# Define color palette for 'M' and 'F'
colors = {'M': '#1f77b4', 'F': '#ff7f0e'}

# Create countplot with hue for gender, explicitly setting colors
ax = sns.countplot(data=df, x='Age Group', hue='Gender', palette=colors)

# Add bar labels (count values)
for bars in ax.containers:
    ax.bar_label(bars, fmt='%d', fontsize=12, label_type='edge', color='black', padding=3)

# Customize plot appearance
ax.set_title('Age Group Distribution by Gender', fontsize=16)
ax.set_xlabel('Age Group', fontsize=14)
ax.set_ylabel('Count', fontsize=14)
ax.tick_params(axis='both', labelsize=12)

# Display the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Aggregate total sales amount by Age Group
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

# Create a larger figure
fig, ax = plt.subplots(figsize=(10, 6))

# Define color palette (each Age Group gets a unique color)
colors = sns.color_palette("viridis", len(sales_age))  # Using a gradient color scheme

# Create the bar plot with assigned colors (explicit hue to avoid warning)
ax = sns.barplot(x='Age Group', y='Amount', data=sales_age, hue='Age Group', palette=colors, width=0.6, legend=False)

# Customize plot appearance
ax.set_title('Total Sales Amount by Age Group', fontsize=16)
ax.set_xlabel('Age Group', fontsize=14)
ax.set_ylabel('Total Amount', fontsize=14)
ax.tick_params(axis='both', labelsize=12)

# Rotate x-axis labels if needed for better readability
plt.xticks(rotation=45)

# Display the plot
plt.show()
```

*From above graphs we can see that most of the buyers are of age group between 26-35 yrs female*

### State

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Group data by State and get the top 10 states with the highest number of orders
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

# Create a larger figure
fig, ax = plt.subplots(figsize=(15, 5))

# Define a color palette for each state
colors = sns.color_palette("coolwarm", len(sales_state))

# Create the bar plot with explicit hue assignment and custom colors
ax = sns.barplot(data=sales_state, x='State', y='Orders', hue='State', palette=colors, width=0.6, legend=False)

# Add bar labels (total number of orders)
for bars in ax.containers:
    ax.bar_label(bars, fmt='%d', fontsize=12, label_type='edge', color='black', padding=3)

# Customize plot appearance
ax.set_title('Total Number of Orders from Top 10 States', fontsize=16)
ax.set_xlabel('State', fontsize=14)
ax.set_ylabel('Total Orders', fontsize=14)
ax.tick_params(axis='both', labelsize=12)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Display the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Group data by State and get the top 10 states with the highest total sales amount
sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

# Create a larger figure
fig, ax = plt.subplots(figsize=(15, 5))

# Define a color palette for each state
colors = sns.color_palette("mako", len(sales_state))

# Create the bar plot with explicit hue assignment and custom colors
ax = sns.barplot(data=sales_state, x='State', y='Amount', hue='State', palette=colors, width=0.6, legend=False)

# Customize plot appearance
ax.set_title('Total Sales Amount from Top 10 States', fontsize=16)
ax.set_xlabel('State', fontsize=14)
ax.set_ylabel('Total Sales Amount', fontsize=14)
ax.tick_params(axis='both', labelsize=12)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Display the plot
plt.show()
```

*From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively*

### Occupation

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Group data by Occupation and sum the total amount
sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

# Set figure size
sns.set(rc={'figure.figsize': (20, 5)})

# Define a color palette for each unique occupation
colors = sns.color_palette("coolwarm", len(sales_state))

# Create the bar plot with explicit hue assignment and no legend
ax = sns.barplot(data=sales_state, x='Occupation', y='Amount', hue='Occupation', palette=colors, dodge=False, legend=False)

# Customize plot appearance
ax.set_title('Total Sales Amount by Occupation', fontsize=16)
ax.set_xlabel('Occupation', fontsize=14)
ax.set_ylabel('Total Sales Amount', fontsize=14)
ax.tick_params(axis='x', rotation=45)  # Rotate x-axis labels for better readability

# Display the plot
plt.show()
```

*From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector*

### Product Category

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set figure size
sns.set(rc={'figure.figsize': (20, 5)})

# Count the number of unique categories and generate a custom color palette (cool tones)
unique_categories = df['Product_Category'].nunique()
colors = sns.color_palette("cool", unique_categories)  # "cool" for blue-green shades

# Create countplot with hue for Product_Category and no legend
ax = sns.countplot(data=df, x='Product_Category', hue='Product_Category', palette=colors, dodge=False, legend=False)

# Add bar labels (count values)
for bars in ax.containers:
    ax.bar_label(bars, fmt='%d', fontsize=10, label_type='edge', color='black', padding=3)

# Customize plot appearance
ax.set_title('Product Category Distribution', fontsize=16)
ax.set_xlabel('Product Category', fontsize=14)
ax.set_ylabel('Count', fontsize=14)
ax.tick_params(axis='x', rotation=45)  # Rotate x-axis labels for readability

# Display the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Group data by Product_Category and sum the total amount, selecting the top 10
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

# Set figure size
sns.set(rc={'figure.figsize': (20, 5)})

# Define a custom color palette for the top 10 categories
colors = sns.color_palette("coolwarm", len(sales_state))

# Create the bar plot with hue for Product_Category and no legend
ax = sns.barplot(data=sales_state, x='Product_Category', y='Amount', hue='Product_Category', palette=colors, dodge=False, legend=False)

# Customize plot appearance
ax.set_title('Top 10 Product Categories by Sales Amount', fontsize=16)
ax.set_xlabel('Product Category', fontsize=14)
ax.set_ylabel('Total Sales Amount', fontsize=14)
ax.tick_params(axis='x', rotation=45)  # Rotate x-axis labels for better readability

# Display the plot
plt.show()
```

*From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category*

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Group data by Product_ID and sum the number of orders, selecting the top 10
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

# Set figure size
sns.set(rc={'figure.figsize': (20, 5)})

# Define a custom color palette for the top 10 Product IDs
colors = sns.color_palette("cubehelix", len(sales_state))

# Create the bar plot with hue for Product_ID and no legend
ax = sns.barplot(data=sales_state, x='Product_ID', y='Orders', hue='Product_ID', palette=colors, dodge=False, legend=False)

# Add bar labels (total number of orders)
for bars in ax.containers:
    ax.bar_label(bars, fmt='%d', fontsize=10, label_type='edge', color='black', padding=3)

# Customize plot appearance
ax.set_title('Top 10 Products by Number of Orders', fontsize=16)
ax.set_xlabel('Product ID', fontsize=14)
ax.set_ylabel('Number of Orders', fontsize=14)
ax.tick_params(axis='x', rotation=45)  # Rotate x-axis labels for better readability

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Group data by Product_ID, sum the number of orders, and select the top 10 most sold products
top_products = df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False)

# Create a larger figure
fig1, ax1 = plt.subplots(figsize=(12, 7))

# Plot the bar chart
top_products.plot(kind='bar', color='#1f77b4', ax=ax1)  # Custom blue color for bars

# Add bar labels
for p in ax1.patches:
    ax1.annotate(f'{p.get_height():,.0f}', (p.get_x() + p.get_width() / 2., p.get_height()), 
                 ha='center', va='bottom', fontsize=10, color='black')

# Customize plot appearance
ax1.set_title('Top 10 Most Sold Products', fontsize=16)
ax1.set_xlabel('Product ID', fontsize=14)
ax1.set_ylabel('Total Number of Orders', fontsize=14)
ax1.tick_params(axis='x', rotation=45)

# Display the plot
plt.show()
```
